```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100,200,300, 500, 800,150 , 12])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c="Black")
```

```python
plt.plot(salario, c="Black", ls="--")
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario, c="Black", ls='--', marker='s', ms=18)
```

```python
salarios_marcos = np.array([100,200,400, 300])
salarios_gileno = np.array([50,300,500, 450])
```

```python
plt.plot(salarios_marcos, c="Black", ls="--", marker="s", ms=8, label="Salrios Marcos")
plt.plot(salarios_gileno, c="Red", ls="--",  marker="s", ms=8, label="Salrios Gileno")
plt.show()
```

```python
plt.plot(salarios_marcos, c="Black", ls="--", marker="s", ms=8, label="Salrios Marcos")
plt.plot(salarios_gileno, c="Red", ls="--",  marker="s", ms=8, label="Salrios Gileno")
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos, c="Black", ls="--", marker="s", ms=8, label="Salrios Marcos")
plt.plot(salarios_gileno, c="Red", ls="--",  marker="^", ms=8, label="Salrios Gileno")
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos, c="Black", ls="--", marker="s", ms=8, label="Salrios Marcos")
plt.plot(salarios_gileno, c="Red", ls="--",  marker="^", ms=8, label="Salrios Gileno")
plt.legend(loc="upper left")
plt.show()
```

```python
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_marcos, c="Black", ls="--", marker="s", ms=10, label="Salrios Marcos")
plt.plot(salarios_gileno, c="Red", ls="--",  marker="^", ms=10, label="Salrios Gileno")
plt.plot(salarios_pedro, c="Green", ls="--",  marker="o", ms=10, label="Salrios Pedro")
plt.legend(loc="upper left")
plt.show()
```
