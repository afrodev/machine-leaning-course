```python
%matplotlib inline
import pandas as pd
import pydataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pydataset.data('titanic', show_doc=True)
```

```python
titanic = pydataset.data('titanic')
```

```python
titanic.head()
```

```python
titanic['class'].value_counts()
```

```python
titanic['class'].value_counts().plot(kind='bar') 
# faz um gr√°fico de barras a partir de um data frame
plt.show()
```

```python
titanic['survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
titanic.groupby('class')['survived'].value_counts().plot(kind='bar') 
# agrupamento
plt.show()
```

```python
estados = pd.read_csv('populacao_brasil.csv')
```

```python
estados.head()
```

```python
estados['total'].hist()
```

```python
fig, ax = plt.subplots()
plt.hist(estados['total'], bins=15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['percent'] = estados['total'] / estados['total'].sum()
```

```python
estados['percent']
```

```python
plt.pie(estados['percent'], labels=estados['estado'])
plt.show()
```

```python
plt.pie(estados['percent'], labels=estados['estado'], autopct='%1.2f%%')
plt.show()
```
