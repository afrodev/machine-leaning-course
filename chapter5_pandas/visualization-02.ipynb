```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 100) # gera 100 valores
```

```python
plt.plot(x, np.sin(x), 'b--') # b-- tracejado azul
plt.plot(x, np.cos(x), 'r--') ## tracejado vermelho
plt.show()
```
